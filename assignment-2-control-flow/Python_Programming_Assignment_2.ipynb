### Example Starter Code


```python
# Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# add Dataset
data = sns.load_dataset("titanic")

# style
sns.set(style="whitegrid")

# 1. 5 Rows
print("First 5 rows of the dataset:")
print(data.head())

# 2. Dataset Info
print("\nDataset Shape:", data.shape)
print("\nDataset Columns:\n", data.columns.tolist())

print("\nSummary Info:")
print(data.info())

print("\nStatistical Summary:")
print(data.describe())

# 3. Missing Values
print("\nMissing Values in Each Column:")
print(data.isnull().sum())


# 4.1 Survival by Passenger Class
plt.figure(figsize=(6,4))
sns.barplot(x="pclass", y="survived", data=data)
plt.title("Survival Rate by Passenger Class")
plt.ylabel("Survival Rate")
plt.xlabel("Passenger Class")
plt.show()

# 4.2 Age Distribution
plt.figure(figsize=(8,5))
sns.histplot(data["age"].dropna(), kde=True, bins=30, color='skyblue')
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# 4.3 Survival Rate by Sex
plt.figure(figsize=(6,4))
sns.barplot(x="sex", y="survived", data=data)
plt.title("Survival Rate by Gender")
plt.ylabel("Survival Rate")
plt.xlabel("Gender")
plt.show()

# 4.4 Survival vs Family Aboard (sibsp + parch)
data["total_family"] = data["sibsp"] + data["parch"]
plt.figure(figsize=(8, 5))
sns.barplot(x="total_family", y="survived", data=data)
plt.title("Survival Rate by Total Family Members on Board")
plt.xlabel("Total Family Members")
plt.ylabel("Average Survival Rate")
plt.show()

# 5. Bonus Visualizations

# 5.1 Pair Plot (selected features)
sns.pairplot(data[["age", "fare", "pclass", "survived"]].dropna(), hue="survived", palette="husl")
plt.suptitle("Pair Plot of Selected Features", y=1.02)
plt.show()

# 5.2 Heatmap of Correlations
plt.figure(figsize=(10,6))
corr = data.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

```
